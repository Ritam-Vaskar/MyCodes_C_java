## 1. If Statement
Executes code only when condition is True

```python
# Example 1: Check if number is positive
number = 15
if number > 0:
    print(f"{number} is a positive number")
    print("This line also executes")
```

## 2. If-Else Statement
Executes one block if condition is True, another if False

```python
# Example 1: Check even or odd
num = 7
if num % 2 == 0:
    print(f"{num} is even")
else:
    print(f"{num} is odd")
```

```python
# Example 2: Login system
password = "python123"
user_input = "python123"

if user_input == password:
    print("Access granted!")
else:
    print("Access denied!")
```

## 3. Elif (Else-If) Statement
Checks multiple conditions in sequence

```python
# Example 1: Grade calculator
marks = 78

if marks >= 90:
    grade = 'A+'
elif marks >= 80:
    grade = 'A'
elif marks >= 70:
    grade = 'B'
elif marks >= 60:
    grade = 'C'
elif marks >= 50:
    grade = 'D'
else:
    grade = 'F'

print(f"Marks: {marks} → Grade: {grade}")
```

## 4. For Loop
Iterates over a sequence (list, tuple, string, range)

```python
# Example 1: Iterate through a list
fruits = ['apple', 'banana', 'cherry', 'mango']
for fruit in fruits:
    print(f"I like {fruit}")
```

```python
# Example 2: Pattern printing
for i in range(1, 6):
    print('* ' * i)
```

## 5. While Loop
Repeats as long as condition is True

```python
# Example 1: Factorial calculator
n = 5
factorial = 1
temp = n

while temp > 0:
    factorial *= temp
    temp -= 1

print(f"{n}! = {factorial}")
```

```python
# Example 2: Reverse a number
num = 12345
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

print(f"Reversed number: {reverse}")
```

## 6. Break and Continue Statements

```python
# Break - Exit loop early
print("Using break:")
for i in range(1, 11):
    if i == 6:
        break
    print(i, end=' ')
print("\nLoop stopped at 6")
```

```python
# Continue - Skip current iteration
print("Using continue (skip even numbers):")
for i in range(1, 11):
    if i % 2 == 0:
        continue
    print(i, end=' ')
print()
```

## 7. Nested Loops and Conditionals

```python
# Example 1: Prime number checker with nested conditionals
def check_prime(n):
    if n <= 1:
        return False
    elif n == 2:
        return True
    else:
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                return False
        return True

# Test prime numbers from 1 to 20
print("Prime numbers from 1 to 20:")
for num in range(1, 21):
    if check_prime(num):
        print(num, end=' ')
print()
```

## 8. Practical Examples

```python
# Example 1: Fibonacci sequence
n = 10
a, b = 0, 1
print(f"First {n} Fibonacci numbers:")
count = 0
while count < n:
    print(a, end=' ')
    a, b = b, a + b
    count += 1
print()
```
