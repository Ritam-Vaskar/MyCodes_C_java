## I. Dictionary Creation & Access using Key, and Updation

```python
student = {"name": "Alice", "age": 20, "grade": "A"}
print(f"Dictionary: {student}")

print(f"\nAccessing by key:")
print(f"Name: {student['name']}")
print(f"Age: {student['age']}")
print(f"Grade: {student.get('grade')}")

student["age"] = 21
print(f"\nAfter updating age: {student}")

student["city"] = "New York"
print(f"After adding city: {student}")
```

## II. Dictionary Length, Type and dict() Constructor

```python
colors = {"red": "#FF0000", "green": "#00FF00", "blue": "#0000FF"}
print(f"Dictionary: {colors}")
print(f"Length: {len(colors)}")
print(f"Type: {type(colors)}")

dict1 = dict(name="Bob", age=25, city="London")
print(f"\nUsing dict() constructor: {dict1}")

dict2 = dict([("a", 1), ("b", 2), ("c", 3)])
print(f"From list of tuples: {dict2}")
```

## III. Dictionary Adding and Removing Items (pop(), popitem()) and Checking

```python
fruits = {"apple": 5, "banana": 3, "cherry": 8, "mango": 2}
print(f"Original: {fruits}")

fruits["orange"] = 6
print(f"\nAfter adding orange: {fruits}")

removed = fruits.pop("banana")
print(f"\nRemoved banana (value={removed}): {fruits}")

last_item = fruits.popitem()
print(f"Removed last item {last_item}: {fruits}")

print(f"\nChecking membership:")
print(f"'apple' in fruits: {'apple' in fruits}")
print(f"'banana' in fruits: {'banana' in fruits}")
```

## IV. Using clear() and del() on Dictionary

```python
dict1 = {"a": 1, "b": 2, "c": 3}
print(f"Before clear: {dict1}")
dict1.clear()
print(f"After clear: {dict1}")

dict2 = {"x": 10, "y": 20, "z": 30}
print(f"\nBefore deleting key: {dict2}")
del dict2["y"]
print(f"After del dict2['y']: {dict2}")

dict3 = {"p": 100, "q": 200}
print(f"\nBefore del: {dict3}")
del dict3
print("After del: dict3 is deleted")

try:
    print(dict3)
except NameError:
    print("Error: dict3 not defined")
```

## V. Looping and Copying Dictionary using copy()

```python
person = {"name": "John", "age": 30, "city": "Paris"}

print("Loop through keys:")
for key in person:
    print(f"{key}: {person[key]}")

print("\nLoop through keys (explicit):")
for key in person.keys():
    print(key, end=' ')

print("\n\nLoop through values:")
for value in person.values():
    print(value, end=' ')

print("\n\nLoop through items:")
for key, value in person.items():
    print(f"{key} = {value}")

copied = person.copy()
print(f"\nOriginal: {person}")
print(f"Copied: {copied}")

copied["age"] = 35
print(f"\nAfter modifying copy:")
print(f"Original: {person}")
print(f"Copied: {copied}")
```

## VI. Nested Dictionary - Creation, Accessing and Looping Items

```python
students = {
    "student1": {"name": "Alice", "age": 20, "grade": "A"},
    "student2": {"name": "Bob", "age": 22, "grade": "B"},
    "student3": {"name": "Charlie", "age": 21, "grade": "A+"}
}

print("Nested Dictionary:")
print(students)

print(f"\nAccessing nested values:")
print(f"Student1 name: {students['student1']['name']}")
print(f"Student2 age: {students['student2']['age']}")
print(f"Student3 grade: {students['student3']['grade']}")

print("\nLooping through nested dictionary:")
for student_id, info in students.items():
    print(f"\n{student_id}:")
    for key, value in info.items():
        print(f"  {key}: {value}")
```

```python
company = {
    "departments": {
        "HR": {"employees": 10, "budget": 50000},
        "IT": {"employees": 25, "budget": 150000},
        "Sales": {"employees": 15, "budget": 80000}
    }
}

print("Company structure:")
for dept, details in company["departments"].items():
    print(f"{dept}: {details['employees']} employees, Budget: ${details['budget']}")
```
