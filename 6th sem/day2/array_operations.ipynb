## I. Array Creation

```python
import array

arr = array.array('i', [10, 20, 30, 40, 50])
print(f"Array: {arr}")
print(f"Type: {type(arr)}")
```

## II. Array Element Access

```python
arr = array.array('i', [10, 20, 30, 40, 50])

print(f"First element: {arr[0]}")
print(f"Last element: {arr[-1]}")
print(f"Slice [1:4]: {arr[1:4]}")
```

## III. Array Length

```python
arr = array.array('i', [10, 20, 30, 40, 50])

print(f"Array: {arr}")
print(f"Length: {len(arr)}")
```

## IV. Looping Array Elements

```python
arr = array.array('i', [10, 20, 30, 40, 50])

print("Using for loop:")
for element in arr:
    print(element, end=' ')
print()

print("\nUsing while loop:")
i = 0
while i < len(arr):
    print(f"Index {i}: {arr[i]}")
    i += 1
```

## V. Adding Array Elements using append()

```python
arr = array.array('i', [10, 20, 30])

print(f"Before: {arr}")
arr.append(40)
arr.append(50)
print(f"After: {arr}")
```

## VI. Removing Array Elements using pop()

```python
arr = array.array('i', [10, 20, 30, 40, 50])

print(f"Before: {arr}")
removed = arr.pop()
print(f"Removed last: {removed}")
print(f"After: {arr}")

removed_at_index = arr.pop(1)
print(f"Removed at index 1: {removed_at_index}")
print(f"Final: {arr}")
```
